```python
from random import randint, choice
import subprocess
import platform
import time
 
monster = False #Global flags 
egg = False
basket = False


class MapGrid:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.walls = []
        #self.start = (0, 0)
        self.goal = (width-1, height-1)
        self.player = self.start = (randint(0, height), randint(1,height))
      
        self.monster = (randint(1,width),randint(1,height)) #Generate the monster in the given grid size
        self.egg = (randint(1,width),randint(1,height))
        self.basket = (randint(1,width),randint(1,height))

    def move_player(self, d): #Define the directions the player can move 
        x = self.player[0]
        y = self.player[1]
        pos = None
         
        
        

        if d[0] == 'r':
            pos = (x + 1, y)
        if d[0] == 'l':
            pos = (x - 1, y)
        if d[0] == 'u':
            pos = (x, y - 1)
        if d[0] == 'd':
            pos = (x, y + 1)

        if pos not in self.walls: #Player has to travel around walls 
            self.player = pos
            
        elif pos in self.monster: #Set flag to true if player and monster have the same position 
            self.player != pos 
            monster = True
            
        elif pos in self.basket: #Set basket flag
            basket = True
            
        elif pos in self.egg: #Set egg flag
            egg = True 
            
    #def move_monster(self, d):
        
        
            
        
            

        elif pos == self.goal:
            print("You made it to the end!")


def draw_grid(g, width=2):
    playing = True 
    for y in range(g.height):
        for x in range(g.width):
            if (x, y) in g.walls:
                symbol = '#'
            elif (x, y) == g.player:
                symbol = 'P'
            elif (x, y) == g.start:
                symbol = 'S'
                
            elif (x, y) == g.goal:
                symbol = '/'
                
            elif (x,y) == g.basket: #Assigning symbols to the different items; characters corresponding the first letter 
                symbol = 'B'
                
            elif (x,y) == g.monster:
                symbol = 'M'
                
            elif (x,y) == g.egg:
                symbol = 'E'
                
            elif (x,y) == g.basket:
                symbol = 'B'
                
            else:
                symbol = '.'
            print("%%-%ds" % width % symbol, end="")
            
            
        print()


def get_walls(g: MapGrid, pct=.10) -> list: #Generate the walls randomly 
        out = []
        for i in range(int(g.height*g.width*pct)//2):

            x = randint(1, g.width-2)
            y = randint(1, g.height-2)

            out.append((x, y))
            out.append((x + choice([-1, 0, 1]), y + choice([-1, 0, 1])))
        return out





def main():
    g = MapGrid(10, 10) #Define the size of the grid
    g.walls = get_walls(g) #Generate the walls 
    
    game_over = False
    playing = True 

    while g.player != g.monster: #g.goal before 
        
        
        
            draw_grid(g)
            d = input("Which way? (r, l, u, d)")
            g.move_player(d)
            
           # print("You made it!")
        
            if monster == True:
                print("Game over! The monster has found you!")
                game_over = True 
                continue 
                
            elif basket == True:
                print("You have found the basket. Now get the eggs!")
                
            elif egg == True:
                print("You have found an egg!")
                
                
                
    again = input("Play Again? [Y/N] ") #Prompt the user for another game
    while again.lower() not in ["y", "n"]: 
        again = input("Please enter Y or N ")
        if again.lower() == "n": #End game
            print("Thanks for playing!")
            playing = False
        else:
            game_over = False #Continue playing



if __name__ == '__main__':
    main()
    
    
```
